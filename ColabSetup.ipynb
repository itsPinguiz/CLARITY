```python
%%bash
set -e

REPO_URL="https://github.com/itsPinguiz/CLARITY" 
BRANCH="main"
TARGET_DIR="/content/progettoLLM"

if [ ! -d "$TARGET_DIR/.git" ]; then
  git clone --depth 1 --branch "$BRANCH" "$REPO_URL" "$TARGET_DIR"
else
  cd "$TARGET_DIR"
  git fetch origin "$BRANCH"
  git reset --hard "origin/$BRANCH"
  git clean -fd
fi
```
