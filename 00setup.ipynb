```python
import os
import sys
import subprocess
from google.colab import drive

# 1. Montaggio di Google Drive
drive.mount('/content/drive', force_remount=True)

DRIVE_WORKSPACE = "/content/drive/MyDrive/progettoLLM"
REPO_DIR = f"{DRIVE_WORKSPACE}/CLARITY"

os.makedirs(DRIVE_WORKSPACE, exist_ok=True)
os.chdir(DRIVE_WORKSPACE)

# 2. Sincronizzazione rigorosa con GitHub
if not os.path.exists(REPO_DIR):
    print("Clonazione del repository in corso...")
    subprocess.run(["git", "clone", "https://github.com/itsPinguiz/CLARITY"], check=True)
else:
    print("Repository rilevato. Allineamento forzato con il branch remoto 'main'...")
    subprocess.run(["git", "fetch", "origin", "main"], cwd=REPO_DIR, check=True)
    subprocess.run(["git", "reset", "--hard", "origin/main"], cwd=REPO_DIR, check=True)
    subprocess.run(["git", "clean", "-fd"], cwd=REPO_DIR, check=True)

# 3. Configurazione del path Python
os.chdir(REPO_DIR)
if REPO_DIR not in sys.path:
    sys.path.insert(0, REPO_DIR)

# 4. Esecuzione dello script di setup per le dipendenze
print("Avvio dello script di setup delle dipendenze (setup.sh)...")
subprocess.run(["bash", "setup.sh"], check=True)

print("Infrastruttura pronta. È ora possibile chiudere questo notebook e aprire quello interno.")
```
